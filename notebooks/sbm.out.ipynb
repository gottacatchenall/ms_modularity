```
function network_df(net)
    sp = species(net)
    df = DataFrame(source=[],target = [])
    for int in SpeciesInteractionNetworks.interactions(net)
        i, j = findfirst(isequal(int[1]), sp), findfirst(isequal(int[2]), sp)
        push!(df.source, i)
        push!(df.target, j)
    end
    df
end
```

```
hadfield_df = network_df(H)
nz_df = network_df(net)

CSV.write("hadfield.csv", hadfield_df)
CSV.write("nz.csv", nz_df)
```
